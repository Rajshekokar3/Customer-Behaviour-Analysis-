```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Allways putt the column name into the snake cases :- lower and underscore
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace("-","_")
```

```python
df.columns
```

```python
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#Create a  colume age_group

labels=['Young Adult','Adult','Middle_aged','Senior']
df['age_grouped']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_grouped']].head(5)
```

```python
df['frequency_of_purchases'].unique()
```

```python
# Purchase  frequency rate
frequency_mapping={
    'Fortnightly':14
    ,'Weekly':7
    ,'Annually':365
    ,'Quarterly':90
    ,'Bi-Weekly':14
    ,'Monthly':30
    ,'Every 3 Months':90}

df['purchase_freuency']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[["purchase_amount","frequency_of_purchases","purchase_freuency"]].head(5)
```
